```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec

### Data
df = pd.read_csv('batch_size.csv')
filtered = df

### Plot

fig = plt.figure(figsize=(10/2.54, 8/2.54))

batch_size = [str(tick) for tick in filtered['batch_size']]

plt.bar(batch_size, filtered['root_mean_squared_error'], color='white', edgecolor='black', linewidth=1.5, zorder=3)
plt.ylabel('Root Mean Squared Error $RMSE$')
plt.xlabel('Batch Size')
plt.grid(True, linestyle='--',color='k', linewidth=1, zorder=0)


# Axis settings
ax = plt.gca()
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

ax.set_ylim(bottom=0, top=35)

# Einheiten auf y-Achse
yunit = '--'
ticks = ax.get_yticks()
ticks = [int(tick) for tick in ticks]
ticks_with_units = [yunit if i == len(ticks) - 2 else ticks[i] for i in range(len(ticks))]
ax.set_yticks(ticks)
ax.set_yticklabels(ticks_with_units)

plt.grid(True, linewidth=1.0)
x = 7.5
plt.text(x, 15, s='Modellparameter:', fontsize=10, fontweight='bold')
plt.text(x, 12, s='Optimierer: Adam', fontsize=10)
plt.text(x, 9, s='Lernrate: 0.005', fontsize=10)
plt.text(x, 6, s='Loss: Mean Squared Error', fontsize=10)
plt.text(x, 3, s='Training: 30 Epochen', fontsize=10)
plt.text(x, 0, s='Batch-Größe: 4', fontsize=10)



plt.savefig('batchsize_selection.pdf', bbox_inches='tight', pad_inches=0.2, format='pdf')
plt.show()
```
