```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec

### Data
data = np.load('00_bis.npz')
bis = data['train'].flatten()

bis[bis == 0.0] = np.nan
bis = (- (bis * 100) + 100)

### Plot

fig = plt.figure(figsize=(19/2.54, 8/2.54))

ax = fig.add_axes([0, 0.1, 0.57, 0.8])
ax = plt.hist(bis, bins=100, density=True, color='black')
plt.ylabel('Auftrittswahrscheinlichkeit')
plt.xlabel('Bispektralindex (BIS)')
plt.grid(True, linestyle='--', linewidth=1)


### Legend and Text
ax_legend = fig.add_axes([0.65, 0.5, 0.2, 0.4])  # [left, bottom, width, height]
ax_text = fig.add_axes([0.85, 0.5, 0.1, 0.4])  # [left, bottom, width, height]

ax_legend.legend()
ax_legend.axis('off')  # Hide the axis for the legend area

ax_text.text(0.5, 0.5, 'Custom Text', va='center', ha='center')
ax_text.axis('off')  # Hide the axis for the text area

plt.tight_layout()
plt.show()
```
