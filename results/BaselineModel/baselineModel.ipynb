```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt

plt.figure(figsize=(15/2.54, 8/2.54))

data = np.load('00_bis.npz')
bis = data['train'].flatten()

print(np.histogram((- (bis * 100) + 100), bins=100, density=True))

bis[bis == 0.0] = np.nan
bis = (- (bis * 100) + 100)

plt.hist(bis, bins=100, density=True, color='black')

#g = sns.displot(data=bis, stat="probability", bins=100)
plt.ylabel('Auftrittswahrscheinlichkeit', fontweight='bold')
plt.xlabel('Bispektralindex (BIS)', fontweight='bold')
plt.grid(True, linestyle='--', linewidth=1)

#plt.savefig('bis_histogram.pdf', bbox_inches='tight', pad_inches=0.2, format='pdf')
plt.show()
```
