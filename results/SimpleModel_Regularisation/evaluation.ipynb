```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd 
import seaborn as sns

#sns.set_theme(rc={'figure.figsize':(15/2.54,8/2.54)})

# Load the data
df = pd.read_csv('regularization.csv')

#print(filtered)


g = sns.barplot(data=df, x='regularization', y='root_mean_squared_error')
plt.ylabel('Root Mean Squared Error', fontweight='bold')
plt.xlabel('Regularisierung', fontweight='bold')
#plt.legend(title='Learning Rate', loc='upper left', ncol=2)
g.grid(True, which='both', linestyle='--', linewidth=1)
g.set_xticklabels(['No Regularisation', 'Batch\nNormalisation', 'Dropout', 'Batch \nNormalisation \n+ Dropout'])

plt.savefig('regularisation_selection.pdf', bbox_inches='tight', pad_inches=0.2, format='pdf')
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.gridspec import GridSpec

### Data
df = pd.read_csv('regularization.csv')
filtered = df


### Plot

fig = plt.figure(figsize=(13/2.54, 8/2.54))

reg = ['No Regu-\nlarisation', 'Batch Norm-\nalisation', 'Dropout', 'Batch \nNormalisation \n+ Dropout']

plt.bar(reg, filtered['root_mean_squared_error'], color='white', edgecolor='black', linewidth=1.5, zorder=3)
plt.ylabel('Root Mean Squared Error $RMSE$')
plt.xlabel('Regularisierung')
plt.grid(True, linestyle='--',color='k', linewidth=1, zorder=0)

# Axis settings
ax = plt.gca()
ax.spines['left'].set_linewidth(1.5)
ax.spines['bottom'].set_linewidth(1.5)

ax.set_ylim(bottom=0, top=10)

# Einheiten auf y-Achse
yunit = '--'
ticks = ax.get_yticks()
ticks = [int(tick) for tick in ticks]
ticks_with_units = [yunit if i == len(ticks) - 2 else ticks[i] for i in range(len(ticks))]
ax.set_yticks(ticks)
ax.set_yticklabels(ticks_with_units)

plt.grid(True, linewidth=1.0)
x = 4
plt.text(x, 4, s='Modellparameter:', fontsize=10, fontweight='bold')
plt.text(x, 3, s='Optimierer: Adam', fontsize=10)
plt.text(x, 2, s='Loss: Mean Squared Error', fontsize=10)
plt.text(x, 1, s='Training: 30 Epochen', fontsize=10)
plt.text(x, 0, s='Batch-Größe: 4', fontsize=10)



plt.savefig('activation_selection.pdf', bbox_inches='tight', pad_inches=0.2, format='pdf')
plt.show()
```
